```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("flights")
df.head()
```

```python
#line plot
plt.plot(df["year"],df["passengers"])
plt.xlabel("year")
plt.ylabel("number of passengers")
plt.title("flight boardings over the years")
plt.show()
plt.show()
```
