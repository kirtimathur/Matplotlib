```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#barplot
```

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP'] #categorical data (a category of language)
students = [23,30,35,40,12] #total number of students enrolled in each course
#find out which course has maximum no. of students 
plt.bar(langs,students)
```

```python
#python has max no. of students
#PHP has min. 
```

```python
fig = plt.figure(dpi=80)
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
plt.bar(langs,students,color ='green',width = 0.6)
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
#comparing year wise sale (not line plot as mostly used to see the trend) w.r.t max and min
```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.bar(year,online_sale,color ='orange',width = 0.6)
plt.xlabel("year")
plt.ylabel("online sale")
plt.title("year wise sales")
plt.show()
```
