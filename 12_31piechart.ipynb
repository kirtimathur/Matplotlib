- pie-chart:
    - used for univariate analysis
    - mostly used for categorical data
    - it tells the distribution of categories of the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
l=[200,300,310,800] #percentage gives better analysis
plt.pie(l);
```

```python
l1=["python","ds","da","st"] #for labels
plt.pie(l,labels=l1)
plt.show()
```

```python
l1=["python","ds","da","st"] #for labels
plt.pie(l,labels=l1,autopct="%0.01f%%")
        #autopct: auto percentage, 0.01f indicates 1 no. after decimal, 0.02 means 2 no.s after decimal
plt.show()
```

```python
l1=["python","ds","da","st"] #for labels
l=[1,1,1,2]
plt.pie(l,labels=l1,explode=l,autopct="%0.01f%%")
plt.show()
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0,0,0,5] ## explode 1 means distance
mycolors=["red","yellow","pink","brown"]

plt.pie(y, labels = mylabels, colors=mycolors,explode = myexplode, shadow=True)
plt.show() 
```
