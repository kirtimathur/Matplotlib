```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
#lineplot using array: arange
```

```python
x=np.arange(1,11)
y=np.arange(12,22)
y=x**3
plt.plot(x,y,color="g",marker="^")
plt.title("scatter plot")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```

```python
#comparisons: multiple lines in one figureÂ¶
```

```python
x1=np.linspace(1,10,5)
x2=np.linspace(10,20,5)
plt.figure(figsize=(6,6))
x3=x2*2
x4=x2*3
plt.plot(x1,x2)
plt.plot(x1,x3)
plt.plot(x1,x4)
plt.show()
```

```python
#adding legend: to give labels/ info for lines
```

```python
plt.figure(figsize=(6,4))
x3=x2*2
x4=x2*3
plt.plot(x1,x2,label="2000")
plt.plot(x1,x3,label="2005")
plt.plot(x1,x4,label="2010")
plt.legend()
plt.show()
```

```python
l=[1,2,3,4,5,6,7] ## product id
price=[400,5000,32002,60005,70089,90000,70064]  ##  total price
plt.plot(l,price,color="r",marker="+")
plt.xlabel("product id")
plt.ylabel("product price")
plt.title("product price comparison")
plt.show()
```

```python
plt.figure(figsize=(6,4),dpi=90) #dpi: dots per inch
l=[1,2,3,4,5,6,7] ## product id
# plt.figure(dpi=200)
online_price=[400,500,322,65,789,90,764]  ##  online price
offline_price=[450,556,400,100,770,110,800] ### offline price
plt.plot(l,online_price,label="online price",color="b",marker="+")
plt.plot(l,offline_price,label="offline price",color="g",marker="^")
plt.legend()
plt.show()
```

```python
# from ablove graph we can see which product is better online/ offline
# 1,2,3 is better online
# 4,5,6,7 price is almost same
```

```python
year=["2013","2014","2015","2016","2017","2018","2019","2020","2021","2022"]
online_sale=[5,8,8,9,10,12,8,7,20,22]
plt.plot(year,online_sale,color="g",marker="o")
plt.xticks(rotation=45) #in x-axis the points are rotated at 45 degree
plt.xlabel("year")
plt.ylabel("sales")
plt.title("trend of sales over the years")
plt.show()
```
