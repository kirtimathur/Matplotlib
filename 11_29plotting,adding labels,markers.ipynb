```python
import matplotlib.pyplot as plt #pyplot is method in matplotlib (plt is alice)
import numpy as np
import pandas as pd
```

```python
l1=[2,3,4] #a seq from linspace, arange, list, tuple,etc
l2=[4,5,6]
plt.plot(l1,l2); #; to remove any extra info above graph or use plt.show()
```

```python
#adding labels
l1=[2,3,4] 
l2=[4,5,6]
plt.plot(l1,l2)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot")
plt.show()
```

```python
#highlighting points : by adding markers

l1=[2,3,4] 
l2=[4,5,6]
plt.plot(l1,l2,marker="^",markersize=18) #diff types, like "o", "^"
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot")
plt.show()
```

```python
#changing line color and style
l1=[2,3,4] 
l2=[4,5,6]
plt.plot(l1,l2,color="cyan",linestyle="--",marker="^",markersize=14)
plt.xlabel("x line")
plt.ylabel("y line")
plt.title("first plot")
plt.show()
```

```python
#lineplot using array: arange
x=np.arange(1,11)
y=np.arange(12,22)
plt.plot(x,y,color="g",marker="^",markersize=14)
plt.title("scatter plot")
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.show()
```
